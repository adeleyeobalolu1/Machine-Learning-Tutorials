```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_data = np.random.random(50)*100
y_data = np.random.random(50)*100

plt.scatter(x_data,y_data, c = "r", marker = "*", s = 150, alpha= 0.3)
plt.show()
```

```python
years = [2006 + x for x in range(16)]
weight = []
for i in range(len(years)):
   weight.append(np.random.randint(71,95))
plt.plot(years,weight, 'r--' )
plt.show()
```

```python
x=['C++','Python','Java','C#','Go']
y = [20,12,45,21,43]

plt.bar(x,y, color = '#abcdef', align ='edge', edgecolor = 'b')
plt.show()
```

```python
ages = np.random.normal(20, 1.5,1000)
plt.hist(ages, bins =10, cumulative=True, edgecolor = 'r')
plt.show()
```

```python
languages = ['C++','Python','Java','C#','Go']
votes = [50,24,14,6,17]
plt.pie(votes, labels = languages, autopct = '%.2f%%', pctdistance = 1.3)
plt.show()
```

```python
years = [2014,2015,2016,2017,2018,2019,2020,2021]
income = [55,56,62,61,72,72,73,75]
income_ticks = list(range(50,81,2))

plt.plot(years,income)
plt.xticks()
plt.yticks(income_ticks,[f'${x}k' for x in income_ticks])
plt.xlabel("Year")
plt.ylabel('Yearly income in USD')
plt.title('Income Growth')
plt.show()
```

```python
stock_a  = [100,102,99,100,101,101,100,102]
stock_b = [90,95,102,104,105,103,109]
stock_c = [110,115,100,105,100,98,95]

plt.style.use('seaborn-v0_8-dark-palette')
plt.plot(stock_a, label = 'Company 1')
plt.plot(stock_b, label = 'Company 2')
plt.plot(stock_c, label = 'Company 3')
plt.legend(loc = 'upper center')
plt.show()
```

```python
x = np.arange(1,100)
plt.figure(figsize=(20,15), dpi = 300)
fig, axs = plt.subplots(2,2)

axs[0,0].plot(x, np.sin(x))
axs[0,0].set_title('Sine Wave')

axs[0,1].plot(x, np.cos(x))
axs[0,1].set_title('Cosine Wave')

axs[1,0].plot(x, np.random.random(99))
axs[1,0].set_title('Random Wave')

axs[1,1].plot(x, np.log(x))
axs[1,1].set_title('Log Wave')

fig.tight_layout()
plt.suptitle('Wave Plots')

plt.show()
```

```python
ax = plt.axes(projection = '3d')
x = np.arange(0,50,0.1)
y = np.sin(x)
z = np.cos(x)

ax.plot(x,y,z)
plt.show()
```

```python
ax = plt.axes(projection = '3d')
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.cos(Y)

fig = plt.figure()

ax.plot_surface(X,Y,Z, cmap = 'Spectral')
ax.view_init(azim = 30, elev= 30)
plt.show()
```
