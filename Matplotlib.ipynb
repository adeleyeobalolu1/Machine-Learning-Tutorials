```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
gas.head()
```

```python
plt.figure(figsize = (8,5))
for country in gas:
  if country != 'Year':
    plt.plot(gas["Year"],gas[country], '.-' ,label = country )

plt.xticks(gas["Year"][::3])
plt.legend()
plt.xlabel("Year")
plt.ylabel("Price in USD", fontdict={"fontweight":'bold','fontsize':18})
plt.title('Gas Prices overtime')
plt.show()
```

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head()
```

```python
bins = list(range(40,110,10))
plt.hist(fifa["Overall"], bins = bins)
plt.xticks(bins)
plt.show()
```

```python
left = fifa.loc[fifa["Preferred Foot"] =='Left'].count()[0]
right = fifa.loc[fifa["Preferred Foot"] =='Right'].count()[0]
labels = ["Right Foot","Left Foot"]
plt.pie([right,left],labels= labels)
plt.show()
```
